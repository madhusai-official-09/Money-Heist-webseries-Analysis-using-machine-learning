```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'MoneyHeist.csv'  
df = pd.read_csv(file_path)
```

```python
print(df)
```

```python
# Display the first few rows of the dataset
print("First 5 rows of the dataset:")
print(df.head())
```

```python
# Display basic information about the dataset
print("\nDataset Information:")
print(df.info())
```

```python
# Display basic statistics of numerical columns
print("\nStatistical Summary:")
print(df.describe())
```

```python
# Data Cleaning (if necessary)
# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Visualize Ratings Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Rate'], bins=10, kde=True)
plt.title('Distribution of Episode Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

```python
# Average Rating per Season
avg_rating_per_season = df.groupby('Season')['Rate'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Season', y='Rate', data=avg_rating_per_season)
plt.title('Average Episode Rating per Season')
plt.xlabel('Season')
plt.ylabel('Average Rating')
plt.grid()
plt.show()
```

```python
# Count of Episodes per Season
episode_count_per_season = df['Season'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
sns.barplot(x=episode_count_per_season.index, y=episode_count_per_season.values)
plt.title('Number of Episodes per Season')
plt.xlabel('Season')
plt.ylabel('Number of Episodes')
plt.grid()
plt.show()
```
